```python
a.Lot_Extent=a.Lot_Extent.fillna(a.Lot_Extent.mean())
```

```python
a.Brick_Veneer_Type=a.Brick_Veneer_Type.fillna("None")
```

```python
a.Basement_Height=a.Basement_Height.fillna("TA")
a.Basement_Condition = a.Basement_Condition.fillna("TA")
a.Exposure_Level=a.Exposure_Level.fillna("No")
a.BsmtFinType1=a.BsmtFinType1.fillna("Unf")
a.Garage_Built_Year=a.Garage_Built_Year.fillna(a.Garage_Built_Year.mean())
a.BsmtFinType2=a.BsmtFinType2.fillna("Unf")
```

```python
a.Brick_Veneer_Area = a.Brick_Veneer_Area.fillna(a.Brick_Veneer_Area.mean())
```

```python
a=a.drop(["Pool_Quality","Fence_Quality","Miscellaneous_Feature","Fireplace_Quality","Lane_Type"],axis=1)
```

```python
a.isnull().sum()
```

```python
a=a.iloc[:,1::]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a[a.select_dtypes(include=["object"]).columns]=a[a.select_dtypes(include=["object"]).columns].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test=train_test_split(a,test_size=0.2)
```

```python
a_train_x=a_train.iloc[:,:-1]
a_train_y=a_train.iloc[:,-1]
```

```python
a_test_x=a_test.iloc[:,:-1]
a_test_y=a_test.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(a_train_x,a_train_y)
```

```python
pred_test=le.predict(a_test_x)
pred_train=le.predict(a_train_x)
```
