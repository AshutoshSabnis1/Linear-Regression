```python
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv("train.csv")
```

```python
a=a.drop("ID",axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a[a.select_dtypes(include=["object"]).columns]=a[a.select_dtypes(include=["object"]).columns].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test=train_test_split(a,test_size=0.2)
```

```python
a_train_x=a_train.iloc[:,1::]
a_train_y=a_train.iloc[:,0]
```

```python
a_test_x=a_test.iloc[:,1::]
a_test_y=a_test.iloc[:,0]
```

```python
from sklearn.linear_model import LinearRegression
l=LinearRegression()
l.fit(a_train_x,a_train_y)
```

```python
Rsq=l.score(a_train_x,a_train_y)
Rsq
```

```python
N=a_train_x.shape[0]
K=a_train_x.shape[1]
adj=1-(1-Rsq)*(N-1)/(N-K-1)
adj
```

```python
l.coef_
```

```python
l.intercept_
```

```python
pred_train=l.predict(a_train_x)
pred_test=l.predict(a_test_x)
```

```python
err_test=a_test_y-pred_test
err_train=a_train_y-pred_train
```

```python
mse_test=np.mean(np.square(err_test))
```

```python
mse_test
```

```python
mse_train=np.mean(np.square(err_train))
mse_train
```

```python
rmse_test=np.sqrt(mse_test)
rmse_test
```

```python
rmse_train=np.sqrt(mse_train)
rmse_train
```

```python
mape=np.mean(np.abs(err_test*100/a_test_y))
```

```python
mape
```

```python
mape=np.mean(np.abs(err_train*100/a_train_y))
mape
```

```python
# We are facing the problem of overfitting
from sklearn.linear_model import Ridge
r=Ridge()
r.fit(a_train_x,a_train_y)
```

```python
r.score(a_train_x,a_train_y)
```

```python
pred_r_train=r.predict(a_train_x)
pred_r_test=r.predict(a_test_x)
```

```python
err_r_train=a_train_y-pred_r_train
err_r_test=a_test_y-pred_r_test
```

```python
mse1=np.mean(np.square(err_r_train))
mse2=np.mean(np.square(err_r_test))
```
