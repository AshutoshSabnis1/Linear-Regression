```python
import numpy as np 
import pandas as pd
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
a=pd.read_excel("titanic.xls")
```

```python
df=a.copy()
```

```python
df.head(30)
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()[df.isnull().sum()*100/df.shape[0]>40]
```

```python
de=["cabin","boat","body","home.dest"]
df=df.drop(columns=de)
```

```python
df.age.value_counts()
```

```python
df.age=df.age.fillna(df.age.mean())
```

```python
df.fare.value_counts()
```

```python
df.fare=df.fare.fillna(df.fare.mean())
```

```python
df=df.drop("name",axis=1)
```

```python
df.embarked.value_counts()
```

```python
df.embarked=df.embarked.fillna("S")
```

```python
df.isnull().sum()
```

```python
# Rearrange columns with 'survived' at index 0
df = df[['survived'] + [col for col in df.columns if col != 'survived']]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
col=df.select_dtypes(include=["object"]).columns
df[col]=df[col].astype(str)
df[df.select_dtypes(include=["object"]).columns]=df[df.select_dtypes(include=["object"]).columns].apply(le.fit_transform)
```

```python
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(df,test_size=0.2)
```

```python
df_train_x=df_train.iloc[:,1:]
df_train_y=df_train.iloc[:,0]
```

```python
df_test_x=df_test.iloc[:,1:]
df_test_y=df_test.iloc[:,0]
```

```python
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
l.fit(df_train_x,df_train_y)
```

```python
pred=l.predict(df_test_x)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix
df_tab=confusion_matrix(df_test_y,pred)
```

```python
df_tab
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(df_test_y,pred)*100
```

```python
# Recall(TPR)
from sklearn.metrics import recall_score
recall_score(df_test_y,pred)*100
```

```python
# Precision
from sklearn.metrics import precision_score
precision_score(df_test_y,pred)*100
```

```python
# F1 Ratio
from sklearn.metrics import f1_score
f1_score(df_test_y,pred)*100
```

```python
pred_prob=l.predict_proba(df_test_x)
pred_prob
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
roc_auc_score(df_test_y,pred)*100
```

```python
pred_prob[:,1]
```

```python
fpr,tpr,ther=roc_curve(df_test_y,pred_prob[:,1])
```
